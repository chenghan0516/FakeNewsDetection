#FakeNewsNet

```
def collect_content(file_path,all_content,dir_path,label):
  all_news = pd.read_csv(file_path)
  for i,news in all_news.iterrows():
    if os.path.isfile("code/fakenewsnet_dataset/" + dir_path + news["id"] + "/news content.json"):
      temp_file = open("code/fakenewsnet_dataset/" + dir_path + news["id"] + "/news content.json","r")
      temp_json = json.load(temp_file)
      temp = pd.DataFrame({"title":[temp_json["title"]],"text":[temp_json["text"]],"label":[label]})
      all_content = all_content.append(temp, ignore_index=True)
      temp_file.close()
  return all_content
```

```
all_content=pd.DataFrame({"title":[],"text":[],"label":[]})
all_content = collect_content("dataset/politifact_real.csv",all_content,"politifact/real/",1)
all_content = collect_content("dataset/politifact_fake.csv",all_content,"politifact/fake/",0)
print(all_content)
all_content.to_csv("politifact.csv",index=False)
```
